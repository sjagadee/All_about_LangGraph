```python
# import libraries
import tools # import the tools (custom tools)
import os

from langchain.agents import create_agent
```

```python
os.environ["LANGCHAIN_PROJECT"] = "Agents LangChain"
```

```python
print(tools.web_search_tool.invoke("Machine Learning"))  # invoke the web search tool
```

## Create Agents in LangChain V1.0

```python
# Create the agent
agent = create_agent(
    model="openai:gpt-4.1-mini",
    tools=[tools.web_search_tool]
)

print("Agent created successfully with model string!")
print("LangChain automatically created ChatOpenAI instance from string")
```

# Test Agent

```python
initial_state = {"messages": "machine learning tutorials"}

response = agent.invoke(initial_state)
for message in response['messages']:
    message.pretty_print()
```
